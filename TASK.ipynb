```python
import pandas as pd
import sqlite3

# Connect to SQLite (or create new DB)
conn = sqlite3.connect('ecommerce.db')

# Load and write CSVs to SQL
files = {
    "ad_sales": "Product_Level_Ad_Sales.csv",
    "total_sales": "Product_Level_Total_Sales.csv",
    "eligibility": "Product_Level_Eligibility.csv"
}

for table, path in files.items():
    df = pd.read_csv(path)
    df.to_sql(table, conn, if_exists="replace", index=False)

print("All tables created in ecommerce.db")
```
